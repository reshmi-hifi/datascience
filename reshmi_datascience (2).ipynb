```python
no_features = ['Data Collection (Start Year)', 'Data Collection (End Year)']
cat_features = [cat for cat in df.columns if cat not in no_features + ['Sr.No.', 'Sex']]
```

```python
fig = px.pie(df, names='Sex')
fig.update_layout(legend_title='Sex', title={'text': 'Distribution of Sex', 'y':0.95, 'x':0.5, 'xanchor': 'center', 'yanchor': 'top'})
fig.show()
```

```python
df['Range (In Year)'] = (df['Data Collection (End Year)'] - df['Data Collection (Start Year)'] + 1).astype('O')
```

```python
df.drop(columns=['Sr.No.'], inplace=True)
```

```python
dy = df[df['Range (In Year)'] == 1]
for m in dy.groupby(['MaritalStatus'], as_index=False)[['MaritalStatus']].value_counts().nlargest(5, columns='count')['MaritalStatus']:
  dm = dy[df['MaritalStatus'] == m].groupby(['Data Collection (End Year)'], as_index=False)[['MaritalStatus']].value_counts().nlargest(5, columns='count')
  fig = px.bar(dm, x='count', y='Data Collection (End Year)', color='Data Collection (End Year)', orientation='h').update_yaxes(type='category', categoryorder='total ascending')
  fig.update_layout(title={'text': f'Top 5 {m} Rate in A Single Year','y':0.95,'x':0.5,'xanchor': 'center','yanchor': 'top'}, xaxis_title='Year')
  fig.show()
```

```python
# %% [code]
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# Create a sample DataFrame
data = {
    'A': [10, 20, np.nan, 40, 50],
    'B': [5, np.nan, 20, 10, 15],
    'C': ['one', 'two', 'three', 'four', 'five'],
    'D': [1, 2, 3, 4, 5]
}

df = pd.DataFrame(data)

# 1. Display the first few rows of the DataFrame
print("First 5 rows of DataFrame:")
print(df.head())

# 2. Check the data types
print("\nData types of each column:")
print(df.dtypes)

# 3. Summary statistics
print("\nSummary statistics:")
print(df.describe())

# 4. Check for missing values
print("\nMissing values in each column:")
print(df.isnull().sum())

# 5. Fill missing values with 0
df.fillna(0, inplace=True)
print("\nData after filling missing values with 0:")
print(df)

# 6. Filter rows where column A is greater than 15
filtered_df = df.query('A > 15')
print("\nFiltered rows where A > 15:")
print(filtered_df)

# 7. Sort values by column B
df_sorted = df.sort_values(by='B', ascending=False)
print("\nData sorted by column B in descending order:")
print(df_sorted)

# 8. Group by column C and calculate the mean for numerical columns
grouped = df.groupby('C').mean()
print("\nGrouped by column C with mean of numerical columns:")
print(grouped)

# 9. Apply a custom function to column D
df['D_squared'] = df['D'].apply(lambda x: x ** 2)
print("\nData after applying a function to column D:")
print(df)

# 10. Rename columns A and B
df.rename(columns={'A': 'Alpha', 'B': 'Beta'}, inplace=True)
print("\nRenamed columns:")
print(df)

# 11. Merge with another DataFrame
df2 = pd.DataFrame({'C': ['one', 'three', 'five'], 'E': [100, 300, 500]})
df_merged = pd.merge(df, df2, on='C')
print("\nMerged DataFrame with df2:")
print(df_merged)

# 12. Concatenate two DataFrames vertically
df_concat = pd.concat([df, df2], axis=0)
print("\nConcatenated DataFrame:")
print(df_concat)

# 13. Pivot table
df_pivot = df.pivot_table(values='Alpha', index='C', aggfunc=np.sum)
print("\nPivot table for column Alpha:")
print(df_pivot)

# 14. Plot a line graph of column D
plt.plot(df['D'], label='D values')
plt.title('Line plot of column D')
plt.legend()
plt.show()

# 15. Create a bar plot of column D
sns.barplot(x=df['C'], y=df['D'])
plt.title('Bar plot of column D')
plt.show()
```
